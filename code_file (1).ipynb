RISK QUANTIFICATION — Real Indicators

```python
def notam_risk_score(notam_list):
    score = 0
    for n in notam_list:
        if "CLOSED" in n["text"]:
            score += 0.8
        if "RESTRICTED" in n["text"]:
            score += 0.6
    return min(score, 1.0)
```
